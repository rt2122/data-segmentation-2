```python
import pandas as pd
from astropy.wcs import WCS
from astropy.io import fits
import numpy as np
from astropy.coordinates import SkyCoord
from astropy import units as u
from matplotlib import pyplot as plt
from skimage.filters import gaussian
from skimage.draw import circle
from tqdm.notebook import tqdm
```

```python
coma_wcs_ind = pd.read_csv('/home/rt2122/Data/coma_wise_idx.csv', index_col='index')
```

```python
coma_wcs_ind.head()
```

```python
coma_wcs_hdul = fits.open('/home/rt2122/Data/fulldepth_neo4_index.fits') 
```

```python
cur_line = coma_wcs_hdul[1].data[19926]
```

```python
cur_line
```

```python
w = WCS(naxis=2)
```

```python
w.wcs.cd = cur_line['CD']
w.wcs.cdelt = cur_line['CDELT']
w.wcs.crpix = cur_line['CRPIX']
w.wcs.crval = cur_line['CRVAL']
w.wcs.ctype = ['RA---TAN', 'DEC--TAN']
w.wcs.lonpole = cur_line['LONGPOLE']
w.wcs.latpole = cur_line['LATPOLE']
w.wcs.set_pv([(0, 0, 0)])
```

```python
coma_hdul = fits.open('/home/rt2122/Data/coma_sweep/south/sweep-190p025-200p030.fits')
```

```python
ra = coma_hdul[1].data['RA']
dec = coma_hdul[1].data['DEC']
```

```python
radec = np.vstack([ra, dec])
radec.shape
```

```python
pixels = w.wcs_world2pix(radec.T, 0)
```

```python
print(pixels[:, 0].min(), pixels[:, 0].max())
print(pixels[:, 1].min(), pixels[:, 1].max())
```

```python
center = (194.9587, 28.0129)
radius = 0.6
```

```python
mask = np.zeros((len(ra)), dtype=np.uint8)
```

```python
sc = SkyCoord(ra=ra*u.degree, dec=dec*u.degree, frame='icrs')
center_sc = SkyCoord(ra=center[0]*u.degree, dec=center[1]*u.degree, frame='icrs')
mask = sc.separation(center_sc).degree < radius
```

```python
np.count_nonzero(mask) / mask.shape[0]
```

```python
print(pixels[mask, 0].min(), pixels[mask, 0].max())
print(pixels[mask, 1].min(), pixels[mask, 1].max())
```

```python
pic = np.zeros((7490, 2773), dtype=np.float64)
```

```python
flux_g = coma_hdul[1].data['FLUX_G']
```

```python
for k in tqdm(range(pixels.shape[0])):
    i, j = pixels[k]
    if i < 0 or j < 0:
        continue
    i = int(i)
    j = int(j)
    pic[i, j] = flux_g[k]
```

```python
fig= plt.figure(figsize=(10,10))
ax= fig.add_axes([0.1,0.1,0.8,0.8], projection=w)

im = ax.imshow(pic, cmap=plt.get_cmap('plasma'), 
               interpolation='nearest', vmin=0, vmax=1)
```

```python
pic_gauss = gaussian(pic, sigma=20)
```

```python
fig= plt.figure(figsize=(10,10))
ax= fig.add_axes([0.1,0.1,0.8,0.8], projection=w)

im = ax.imshow(pic_gauss, cmap=plt.get_cmap('plasma'), 
               interpolation='nearest', vmin=0, vmax=4)
```

```python
pic_circles = np.zeros((1571, 1571), dtype=np.float64)
for k in tqdm(range(pixels.shape[0])):
    if mask[k]:
        i, j = pixels[k]
        i = int(i)
        j = int(j)
        r = np.log(np.abs(flux_g[k])) / 1.5
        coords = circle(i, j, r, shape=pic_circles.shape) 
        pic_circles[coords] += 1
```

```python
fig= plt.figure(figsize=(10,10))
ax= fig.add_axes([0.1,0.1,0.8,0.8])

im = ax.imshow(pic_circles, cmap=plt.get_cmap('plasma'), 
               interpolation='nearest', vmin=0, vmax=1)
```

```python
pic_circles_gauss = gaussian(pic_circles, 1)
```

```python
fig= plt.figure(figsize=(10,10))
ax= fig.add_axes([0.1,0.1,0.8,0.8])

im = ax.imshow(pic_circles_gauss, cmap=plt.get_cmap('plasma'), 
               interpolation='nearest', vmin=0, vmax=1)
```
