```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
a = np.load('/home/rt2122/Data/Planck/pregen/4/pic.npy')
```

```python
def draw_sqaure(a, x, y, th=3, num=None):
    if num is None:
        num = a.max() + 10
    a[x:x+64+th,y:y+th, :] = num
    a[x:x+64+th,y+64:y+64+th, :] = num
    
    a[x:x+th,y:y+64+th, :] = num
    a[x+64:x+64+th,y:y+64+th, :] = num
```

```python
plt.figure(figsize=(10,10))
cur = np.copy(a)
draw_sqaure(cur, 0, 0, num=0)
plt.imshow(cur[:256,:256,:3])
```
