```python
import numpy as np
import pandas as pd
import healpy as hp
from astropy.coordinates import SkyCoord
from astropy import units as u
from DS_healpix_fragmentation import one_pixel_fragmentation
```

```python
nside=2**5
```

```python
a = np.arange(hp.nside2npix(nside))
hp.mollview(a, nest=True)
```

```python
l, b = hp.pix2ang(nside=nside, ipix=a, lonlat=True, nest=True)
matched = np.logical_and(l <= 180, np.abs(b) >= 20)
mask = np.zeros(hp.nside2npix(nside))
mask[matched]=1
hp.mollview(mask, nest=True)
```
