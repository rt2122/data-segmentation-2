```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from DS_data_transformation import colors_iterator
```

```python
def z_m500_max_pred(df, alpha=0.7):
    
    df = df[df['status'] != 'fp']
    df = df[df['z'] != -1]
    df.index = np.arange(len(df))
    
    fig = plt.figure(figsize=(16, 10))
    ax = fig.add_subplot(projection='3d')
    
    color_it = colors_iterator()
    
    cats = set(df['catalog'])
    for cat in cats:
        cur = df[df['catalog'] == cat]
        s = ax.scatter(cur['z'], cur['M500'], cur['max_pred'], c=next(color_it), alpha=alpha)
        s.set_label(cat)
    ax.set_xlabel('z')
    ax.set_ylabel('M500')
    ax.set_zlabel('max_pred')
    ax.legend()
```

```python
z_m500_max_pred(pd.read_csv('/home/rt2122/Data/detected_cats/full_pz_all_found34_thr0.1_step8_no_fn.csv'))
```
