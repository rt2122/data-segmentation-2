```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

```python
pz_rot = pd.read_csv('/home/rt2122/Data/detected_cats/full_pz_rot28_thr0.1_step8_no_fn.csv')
pz_all = pd.read_csv('/home/rt2122/Data/detected_cats/full_pz_all_found34_thr0.1_step8_no_fn.csv')
pz_rot = pz_rot[pz_rot['status'] == 'tp']
pz_rot = pz_rot[pz_rot['M500'] != 0]
pz_rot.index = np.arange(len(pz_rot))
pz_all = pz_all[pz_all['status'] == 'tp']
pz_all = pz_all[pz_all['M500'] != 0]
pz_all.index = np.arange(len(pz_all))
```

```python
_, ax = plt.subplots(1, 1, figsize=(10, 10))
s = ax.scatter(pz_rot['M500'], pz_rot['max_pred'], alpha=0.5)
s.set_label('pz_rot')
ax.set_ylabel('max_pred')
ax.set_xlabel('M500')
ax.legend()
```

```python
_, ax = plt.subplots(1, 1, figsize=(10, 10))
s = ax.scatter(pz_all['M500'], pz_all['max_pred'], alpha=0.5)
s.set_label('pz_all')
ax.set_ylabel('max_pred')
ax.set_xlabel('M500')
ax.legend()
```
