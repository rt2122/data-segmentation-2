```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import healpy as hp
import os
```

```python
dirname = '/home/rt2122/Data/Planck/normalized/'
files = sorted(next(os.walk(dirname))[-1], key=lambda x:int(x[:3]))
files
```

```python
planck = {}
for file in files:
    planck[int(file[:3])] = np.load(os.path.join(dirname, file))
```

```python
plt.figure(figsize=(10,6), num=0)
hp.mollview(planck[100], fig=0, nest=True, norm='hist', title='Planck HFI 100 GHz')
```
