```python
def plot_history(hist_file, loss_tick=None, cut_eps=[0, -1]):
    import pickle
    import pandas as pd
    import numpy as np
    from matplotlib import pyplot as plt
    
    hist = None
    with open(hist_file, 'rb') as f:
        hist = pickle.load(f)
        hist = pd.DataFrame(hist, index=np.arange(1, len(hist['loss']) + 1))
    _, ax = plt.subplots(2, 1, figsize=(10,14))
    
    hist = hist.iloc[cut_eps[0]:cut_eps[1]]
    
    line, = ax[0].plot(hist.index, hist['loss'], 'co-')
    line.set_label('loss')
    line, = ax[0].plot(hist.index, hist['val_loss'], 'co-', alpha=0.5)
    line.set_label('val_loss')
    
    for c, metr in zip('rb', ['iou', 'dice']):
        line, = ax[1].plot(hist.index, hist[metr], c+'o-')
        line.set_label(metr)
        line, = ax[1].plot(hist.index, hist['val_'+metr], c+'o-', alpha=0.5)
        line.set_label('val_'+metr)
    
    for i in range(2):
        ax[i].legend()
        ax[i].set_xticks(hist.index, minor=True)
        ax[i].set_xticks(hist.index[4::5])
        ax[i].grid(True, axis='both', which='major')
        ax[i].grid(True, axis='both', which='minor', alpha=0.2)
    if not (loss_tick is None):
        ax[0].set_yticks(loss_tick[::5])
        ax[0].set_yticks(loss_tick, minor=True)
```

```python
plot_history('/home/rt2122/Models/pz_only_rot/pz_only_rot.ep1-50.history')
```

```python
def m500_z_graph(cat_name, name):
    df = pd.read_csv(cat_name)
    df = df[df['status'] != 'fn']
    df.index = np.arange(len(df))
    
    pz = pd.read_csv('/home/rt2122/Data/clusters/planck_z.csv')
    mcxc = pd.read_csv('/home/rt2122/Data/original_catalogs/csv/MCXC.csv')
    act = pd.read_csv('/home/rt2122/Data/original_catalogs/csv/ACT.csv')
    
    mcxc_sc = SkyCoord(ra=np.array(mcxc['RA']) * u.degree, dec=np.array(mcxc['DEC']) * u.degree, frame='icrs')
    act_sc = SkyCoord(ra=np.array(act['RA']) * u.degree, dec=np.array(act['DEC']) * u.degree, frame='icrs')
    sc = SkyCoord(ra=np.array(df['RA']) * u.degree, dec=np.array(df['DEC']) * u.degree, frame='icrs')
    
    idx, d2d, _ = sc.match_to_catalog_sky(mcxc_sc)
    matched = d2d.degree < (400 / 60)
    mcxc_matched = df[matched].copy()
    mcxc_matched.index = np.arange(len(mcxc_matched))
    mcxc_matched['z'] = np.array(mcxc['z'].iloc[idx[matched]])
    mcxc_matched['M500'] = np.array(mcxc['M500'].iloc[idx[matched]])
    idx, d2d, _ = sc.match_to_catalog_sky(act_sc)
    matched = d2d.degree < (400 / 60)
    act_matched = df[matched].copy()
    act_matched.index = np.arange(len(act_matched))
    act_matched['z'] = np.array(act['z'].iloc[idx[matched]])
    act_matched['M500'] = np.array(act['M500'].iloc[idx[matched]])
    
    _, ax = plt.subplots(1, 3, figsize=(30, 10), sharex=True, sharey=True)
    for i, (st, en) in enumerate([[0, 1], [0, 0.4], [0.7, 1]]):
        max_pred_cut = np.logical_and(mcxc_matched['max_pred'] >= st, mcxc_matched['max_pred'] <= en)
        s = ax[i].scatter(mcxc_matched[max_pred_cut]['z'], mcxc_matched[max_pred_cut]['M500'], c='r', alpha=0.1)
        s.set_label('MCXC matched with {name} in max_pred [{st}, {en}]'.format(st=st, en=en, name=name))
        
        max_pred_cut = np.logical_and(act_matched['max_pred'] >= st, act_matched['max_pred'] <= en)
        s = ax[i].scatter(act_matched[max_pred_cut]['z'], act_matched[max_pred_cut]['M500'], c='c', alpha=0.1)
        s.set_label('act matched with {name} in max_pred[{st}, {en}]'.format(st=st, en=en, name=name))
        
        s = ax[i].scatter(pz['z'], pz['M500'], c='b', alpha=0.1)
        s.set_label('planck_z')
        
        ax[i].set_xlabel('z')
        ax[i].set_ylabel('M500')
        ax[i].grid()
        ax[i].legend()
```

```python
m500_z_graph('/home/rt2122/Data/detected_cats/full_pz_rot28_thr0.1_step8_no_fn.csv', 'pz_rot28')
```
