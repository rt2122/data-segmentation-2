```python
import pandas as pd
import numpy as np
from DS_models_stat import cut_cat, make_histogram
from matplotlib import pyplot as plt
from DS_data_transformation import Counts_cat, found_all_error_calculated
from tqdm.notebook import tqdm
import os
%config InlineBackend.figure_format = 'retina'
```

```python
dirname = '/home/rt2122/Data/original_catalogs/csv/'
true_cats = next(os.walk(dirname))[-1]
true_cats = {os.path.splitext(os.path.basename(s))[0] : os.path.join(dirname, s) for s in true_cats}
true_cats['PSZ2(z)'] = '/home/rt2122/Data/clusters/planck_z.csv'
true_cats['eROSITA'] = '/home/rt2122/Data/SRGz/clusters/clusters1_b20_edit.csv'
det_cat = '/home/rt2122/Data/detected_cats/full_pz_all_found34_thr0.1_step8_no_fn.csv'
```

```python
def a1(ax):
    df = pd.read_csv('/home/rt2122/Data/detected_cats/full_pz_all_found34_thr0.1_step8_no_fn.csv')
    df = cut_cat(df, big_pix=[41])
    bins = np.arange(0.1, 1.1, 0.1)
    vals = []
    for st, en in zip(bins[:-1], bins[1:]):
        cur_df = cut_cat(df, dict_cut={'max_pred' : [st, en]})
        vals.append(len(cur_df))
    make_histogram(ax, counts_list=[vals], bins=bins, label_list=['SZcat(AL) 41'])
```

```python
counts = {}
n_bins = 10
for cat_name in tqdm(['MCXC', 'ACT', 'PSZ2(z)']):
    cc = found_all_error_calculated(cut_cat(pd.read_csv(det_cat), big_pix=[41]), 
                                    cut_cat(pd.read_csv(true_cats[cat_name]), big_pix=[41]), n_bins=n_bins, name=cat_name) 
    counts[cat_name] = cc
```

```python
def a2(ax, counts):
    counts_list = []
    cats = list(counts.keys()) 
    for cat in cats:
        counts_list.append(counts[cat].found_counts)
    bins = counts[cats[0]].bins
    make_histogram(ax, counts_list=counts_list, bins=bins, label_list=cats)
```

```python
_, ax = plt.subplots(1, 1, figsize=(10 * 1, 10), sharey=True)
a1(ax[0])
a2(ax[1], counts)
```

```python
n_bins = 10
c_erosita = found_all_error_calculated(cut_cat(pd.read_csv(det_cat), dict_cut={'l' : [0, 180], 'b' : [20, np.inf]}), 
                                pd.read_csv(true_cats['eROSITA']), n_bins=n_bins, name=cat_name) 
```

```python
def a3(ax):
    df = pd.read_csv('/home/rt2122/Data/detected_cats/full_pz_all_found34_thr0.1_step8_no_fn.csv')
    df = cut_cat(df, dict_cut={'l' : [0, 180], 'b' : [20, np.inf]})
    bins = np.arange(0.1, 1.1, 0.1)
    vals = []
    for st, en in zip(bins[:-1], bins[1:]):
        cur_df = cut_cat(df, dict_cut={'max_pred' : [st, en]})
        vals.append(len(cur_df))
    make_histogram(ax, counts_list=[vals], bins=bins, label_list=['SZcat(AL) eb20'])
```

```python
_, ax = plt.subplots(1, 2, figsize=(10 * 2, 10))
a3(ax[0])
a2(ax[1], {'eROSITA' : c_erosita})
```
