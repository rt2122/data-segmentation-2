```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from DS_Planck_Unet import load_planck_model
from DS_detector import gen_pics_for_detection, connect_masks
import DS_Coma_Cluster_param as cc_prm
from DS_healpix_fragmentation import radec2pix, one_pixel_fragmentation, \
    matr2dict, draw_proper_circle, draw_proper_circle_s, zoom_to_circle
```

```python
pz_model = load_planck_model(
    '/home/rt2122/Models/planck_z/f8d2.ep0014-vl0.006570-l0.004067.hdf5')
big_pix = radec2pix(cc_prm.ra_center, cc_prm.dec_center, 2)
all_dict = gen_pics_for_detection(big_pix, pz_model, step=8)
big_ans = connect_masks(all_dict['ans'], all_dict['pic_idx'])
big_pic = np.load('/home/rt2122/Data/Planck/pregen/{pix}/pic.npy'.format(pix=big_pix))
```

```python
matr = one_pixel_fragmentation(2, big_pix, 10)
mdict = matr2dict(matr)
mask = draw_proper_circle(cc_prm.ra_center, cc_prm.dec_center, cc_prm.radius, 2048, 
                         mdict, (1024, 1024, 1), coords_mode=True)
ans = zoom_to_circle(mask, big_ans) 
pic = zoom_to_circle(mask, big_pic)
```

```python
gpic = draw_proper_circle_s(np.array(cc_prm.obj)[:, 0], np.array(cc_prm.obj)[:, 1], 
                           0.01, 2048, 
                         mdict, (1024, 1024, 1))
gpic = zoom_to_circle(mask, gpic) 
```

```python
plt.figure(figsize=(10, 8))
plt.imshow(np.dstack([gpic, pic[:, :, 0], ans[:,:,0] >= 0.1]))
```
