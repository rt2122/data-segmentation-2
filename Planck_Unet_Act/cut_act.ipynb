```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
act = pd.read_csv('/home/rt2122/Data/clusters/act.csv')
planck_z = pd.read_csv('/home/rt2122/Data/clusters/planck_z.csv')
mcxcwp = pd.read_csv('/home/rt2122/Data/clusters/mcxcwp.csv')
```

```python
_, ax = plt.subplots(figsize=(8, 6))
line = ax.scatter(planck_z['z'], planck_z['M500'], c='b', alpha=0.3)
line.set_label('planck_z')
line = ax.scatter(mcxcwp['z'], mcxcwp['M500'], c='r', alpha=0.3)
line.set_label('mcxcwp')
line = ax.scatter(act['z'], act['M500'], c='c', alpha=0.1)
line.set_label('act')
ax.set_xlabel('z')
ax.set_ylabel('M500')
ax.plot([0, 1.5], [0, 4])
ax.legend()
#y>=8x/3
```

```python
act_cut = act[act['M500'] >= (8 * act['z'] / 3)]
len(act), len(act_cut)
```

```python
_, ax = plt.subplots(figsize=(8, 6))
line = ax.scatter(planck_z['z'], planck_z['M500'], c='b', alpha=0.3)
line.set_label('planck_z')
line = ax.scatter(mcxcwp['z'], mcxcwp['M500'], c='r', alpha=0.3)
line.set_label('mcxcwp')
line = ax.scatter(act_cut['z'], act_cut['M500'], c='c', alpha=0.1)
line.set_label('act')
ax.set_xlabel('z')
ax.set_ylabel('M500')
ax.plot([0, 1.5], [0, 4])
ax.legend()
#y>=8x/3
```
