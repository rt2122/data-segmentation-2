```python
from DS_detector import gen_pics_for_detection, detect_clusters
from DS_Planck_Unet import *
from DS_data_transformation import draw_df
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from tqdm.notebook import tqdm
from tensorflow import convert_to_tensor
```

```python
thr_list = [0.1 * i for i in range(9, 0, -1)]
```

```python
epoch_dict = {}
for ep, file in tqdm([
    (1, '/home/rt2122/Models/planck_z/f8d2.ep0001-vl0.024432-l0.121729.hdf5'),
    (5, '/home/rt2122/Models/planck_z/f8d2.ep0004-vl0.008401-l0.008484.hdf5'),
    (8, '/home/rt2122/Models/planck_z/f8d2.ep0008-vl0.006764-l0.005214.hdf5'),
    (14, '/home/rt2122/Models/planck_z/f8d2.ep0014-vl0.006570-l0.004067.hdf5'),
    (24, '/home/rt2122/Models/planck_z/f8d2.ep0024-vl0.006689-l0.003243.hdf5'),
    (28, '/home/rt2122/Models/planck_z/f8d2.ep0028-vl0.006358-l0.003048.hdf5'),
    (40, '/home/rt2122/Models/planck_z/f8d2.ep0040.hdf5')
        ]):
    model = load_planck_model(file) 
    all_dict = gen_pics_for_detection(6, model, step=16)
    
    stat_p = []
    stat_pnz = []
    stat_m = []
    for thr in tqdm(thr_list):
        stat_p.append(pd.DataFrame(detect_clusters(all_dict, thr, main_cat='planck_z', 
                                    max_dist=5/60), index=[0]))
        stat_pnz.append(pd.DataFrame(detect_clusters(all_dict, thr, main_cat='planck_no_z', 
                                    max_dist=5/60), index=[0]))
        stat_m.append(pd.DataFrame(detect_clusters(all_dict, thr, main_cat='mcxcwp', 
                                    max_dist=5/60), index=[0]))
    stat_p = pd.concat(stat_p)
    stat_pnz = pd.concat(stat_pnz)
    stat_m = pd.concat(stat_m)
    epoch_dict[ep] = {'p':stat_p, 'pnz':stat_pnz, 'm':stat_m}
```

```python
def draw_df_new(data, base, figsize=(8, 6), comment='comment',min_alpha=0.1, 
                grids=True, xgrid=None, ygrid=None,
               save=False, invert=True):
    from matplotlib import pyplot as plt
    import numpy as np
    fig, ax = plt.subplots(1, figsize=figsize)
    colors = 'bgrcmykw'
    markers = '.ov<>*hx'
    if invert:
        plt.gca().invert_xaxis()
    max_y = 0
    min_y = 100
    alpha = np.arange(min_alpha, 1.0, (1.0 - min_alpha) / (len(data) + 1))
    for i in range(len(data)):
        label = list(data.keys())[i]
        line, = ax.plot(base, data[label], colors[i]+markers[i]+'-', alpha=alpha[i])
        line.set_label(label)
        max_y = max(max_y, max(data[label]))
        min_y = min(min_y, min(data[label]))
    ax.legend()
    if grids:
        if xgrid is None:
            ax.set_xticks(base)
        else:
            ax.set_xticks(xgrid)
            
        if ygrid is None:
            ax.set_yticks(np.arange(min_y, max_y, (max_y-min_y) / 10))
        else:
            ax.set_yticks(ygrid)
        #ax.grid(True)
        plt.grid(b=True, which='major', color='#666666', linestyle=':')
    plt.text(0.6, 0.1, comment)
    if save:
        plt.savefig('pic.png')
    plt.show()
```

```python
draw_df_new({epoch : epoch_dict[epoch]['p']['tp']/
             (epoch_dict[epoch]['p']['tp']+epoch_dict[epoch]['p']['fn'])
             for epoch in epoch_dict},
        thr_list, ygrid=np.arange(0,1.1, 0.1), comment='planck_z recall')
```
