```python
def interopalte_planck(mdict, matr, nside, fits_name, radius, inclusive=True):
    from astropy.coordinates import SkyCoord
    from astropy import units as u
    from astropy.io import fits
    from tqdm.notebook import tqdm
    import healpy as hp
    
    data = None
    with fits.open(fits_name) as hdul:
        data = np.array(hdul[1].data['I_STOKES'])
    planck_nside = 2048
    
    pic = np.zeros(matr.shape)
    for x in tqdm(range(pic.shape[0])):
        vec = hp.pix2vec(nside=nside, ipix=matr[x].flatten(), nest=True)
        vec = np.array(vec)
        for y in range(pic.shape[1]):
            disc = hp.query_disc(planck_nside, vec[:, y], np.radians(radius), 
                                 nest=True, inclusive=inclusive)
            pic[x, y] = np.mean(data[disc])
    return pic
```

```python
def draw_circles(ra, dec, nside, shape, mdict, data, coef=0.02):
    from skimage.draw import circle
    
    pic = np.zeros(shape)
    coef = shape[0] * coef / max(data) 
    pix = radec2pix(ra, dec, nside)
    coords = [(*mdict[pix[i]], data[i]) for i in range(data.shape[0]) 
              if pix[i] in mdict]
    for x, y, flux in coords:
        pic[circle(x, y, coef*flux, shape=shape)] = 1
    return pic
```

```python
def pic2fits1(pic, wcs, fitsname):
    from astropy.io import fits
    import numpy as np

    hdul = fits.HDUList([fits.PrimaryHDU(), 
        fits.ImageHDU(np.stack([pic[:,:,i] for i in range(pic.shape[-1])]))])

    hdul.writeto(fitsname, overwrite=True)
```
