```python
from astropy.coordinates import SkyCoord
from astropy import units as u
import numpy as np
import pandas as pd
from DS_data_transformation import *
import DS_Coma_Cluster_param as cc_prm
from DS_WCS_projections import *
from skimage.filters import gaussian
```

```python
df = fits2df('/home/rt2122/Data/coma_sweep/north/sweep-190p025-200p030.fits')
```

```python
sc = SkyCoord(ra=np.array(df['RA'])*u.degree, 
             dec=np.array(df['DEC'])*u.degree, frame='icrs')
sc_cen = SkyCoord(ra=cc_prm.ra_center*u.degree, 
                 dec=cc_prm.dec_center*u.degree, frame='icrs')
in_rad = sc.separation(sc_cen).degree < 1.1
np.count_nonzero(in_rad)
```

```python
df = df[in_rad]
df.index = np.arange(df.shape[0])
```

```python
wcs = custom_wcs(cc_prm.ra_center, cc_prm.dec_center)
pixels = wcs.all_world2pix(np.array(df['RA']), np.array(df['DEC']), 0)
pixels = np.array(pixels, dtype=np.int32).T
```

```python
inpic = 0
```

```python
pic = np.zeros((2048, 2048, 3))
for i in range(pixels.shape[0]):
    x, y = pixels[i]
    if x >= 0 and x < 2048 and y >= 0 and y < 2048:
        for k, ch in enumerate(['FLUX_G', 'FLUX_R', 'FLUX_Z']):
            pic[x, y, k] = df[ch][i]
        inpic += 1
    else:
        print(x, y)
```

```python
show_pic(pic, projection=wcs)
```
