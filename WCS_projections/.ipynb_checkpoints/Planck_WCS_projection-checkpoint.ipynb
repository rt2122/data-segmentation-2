```python
from astropy.io import fits
import numpy as np
from DS_WCS_projections import *
import DS_Coma_Cluster_param as cc_prm
import os
from tqdm.notebook import tqdm
from DS_healpix_fragmentation import radec2pix
from matplotlib import pyplot as plt
```

```python
c_wcs = custom_wcs(cc_prm.ra_center, cc_prm.dec_center)
```

```python
planck_pic = np.dstack([np.zeros(c_wcs.array_shape) for i in range(6)])
```

```python
dirname = '/home/rt2122/Data/Planck/'
files = sorted(next(os.walk(dirname))[-1])
files
```

```python
matr_pixels = np.zeros(c_wcs.array_shape, dtype=np.int32)
```

```python
planck_nside = 2048
```

```python
for i in tqdm(range(matr_pixels.shape[0])):
    ra, dec = c_wcs.all_pix2world(np.full(matr_pixels.shape[0], i, dtype=np.int32), 
                                  np.arange(matr_pixels.shape[1], dtype=np.int32), 0)
    matr_pixels[i] = radec2pix(ra, dec, planck_nside)
```

```python
for i_f, file in enumerate(files):
    i_st = None
    with fits.open(os.path.join(dirname, file)) as hdul:
        i_st = np.array(hdul[1].data['I_STOKES'])
    for i in tqdm(range(matr_pixels.shape[0])):
        for j in range(matr_pixels.shape[1]):
            planck_pic[i, j, i_f] = i_st[matr_pixels[i, j]]
```

```python
a = np.zeros((4, 4))
#     x, y
c = np.array([5, 2, 8, 90, 1, 5])
a[1] = c[[3, 4, 1, 0]]
a
```

```python
show_pic(planck_pic[:,:,:3], projection=c_wcs)
```

```python
matr_pixels = matr_pixels.astype(np.float32)
matr_pixels -= matr_pixels.min()
matr_pixels[matr_pixels==0] = 0.001
matr_pixels /= matr_pixels.max()
```
