```python
import pandas as pd
import numpy as np
import DS_Coma_Cluster_param as cc_prm
from DS_WCS_projections import *
from DS_data_transformation import show_pic, n_max_flux_df
from skimage.filters import gaussian
from matplotlib import pyplot as plt
```

```python
def orig_wcs(ra, dec):
    from astropy.io import fits
    from astropy.wcs import WCS

    cutout_url = 'https://www.legacysurvey.org/viewer/cutout.fits?ra={:.4f}&dec={:.4f}&layer=sdss2&pixscale=4.00'
    w = None
    with fits.open(cutout_url.format(ra, dec)) as hdul:
         w = WCS(hdul[0].header)
    return w
```

```python
wcs = orig_wcs(cc_prm.ra_center, cc_prm.ra_center)
```

```python
wcs = custom_wcs(cc_prm.ra_center, cc_prm.dec_center)
```
